````python
import pandas as pd
import numpy as np
change_scale = np.vectorize(lambda y: np.power(2,y)-1)
````

````python
metadata = pd.read_csv("../../data/GBM_Neftel/Metadata.txt", skiprows=[1], index_col=0, delimiter="\t")
metadata
````

````python
dataset = pd.read_csv("../../data/GBM_Neftel/IDHwtGBM.processed.SS2.logTPM.txt", index_col=0, delimiter="\t")
````

````python
dataset
````

````python
# Group cells from all patients by cell type (malignant cells grouped as one type)
#metadata_tme = metadata[metadata["CellAssignment"]!="Malignant"]
cts = {}
for a,b in metadata.groupby("CellAssignment"):
    cts[a] = b
````

````python
# For each group, calculate average gene expression
gexp = pd.DataFrame()

for ct in cts:
    # for example: take all vectors of gene expression of oligodendrocytes and
    cells_ct = dataset[cts[ct].index].apply(change_scale)
    
    # calculate mean from them (sum(oligo_cells) / #oligo_cells)
    # This is the source gene expression vector for the cell type
    gexp[ct] = cells_ct.agg(np.mean, axis=1)
````

````python
marker_genes = {}
for ct in gexp:
    gexp_ct = gexp[ct]
    threshold = gexp_ct.quantile(0.75) # 75th percentile
    candidates = gexp[gexp[ct]>=threshold]
    #candidates = gexp_ct.loc[lambda x: x >= threshold]
    #max_genes = gexp.loc[:, gexp.columns != ct]
    candidates["max_others"] = candidates.loc[:, candidates.columns != ct].agg(np.max, axis=1)
    result = (candidates[ct] - candidates["max_others"])/candidates["max_others"]
    marker_genes[ct] = result.sort_values(ascending=False)
````

````python
marker_genes
````

````python
for k in marker_genes:
    if k == "Malignant": continue
    marker_genes[k].to_csv(f"../../data/GBM_Neftel/signatures/{k}.csv")
````
